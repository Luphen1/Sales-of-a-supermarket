## Convert column Admission Type  and Discharge Date  to data type.

```python
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"], format = "%Y-%m-%d")
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"], format = "%Y-%m-%d")
#Check df
df.info()
```

## Create column Age_group

```python
def Age_Group(Age):
    if Age <= 19:
        return "Teenager"
    elif Age <= 29:
        return "Youth"
    elif Age <= 39:
        return "Young Adult"
    elif Age <= 49:
        return "Adult"
    elif Age <= 59:
        return "Middle Aged Adult"
    elif Age <= 69:
        return "Senoir"
    elif Age <= 85:
        return "Elderly"
    else:
        return "Age out of range"
df["Age_Group"] = df['Age'].apply(Age_Group)
df.head(5)
```

#### 1.What age group had the higest intake of billing cost?

```python
#Convert column Billing Amount to float data type.
def coverted_billing_amount(df):
    Billing_amount = df['Billing Amount'] = df['Billing Amount'].apply(lambda x: float(f"{x:.2f}"))
    return Billing_amount

result = coverted_billing_amount(df)
print(result)
```

```python
def age_group_stat(df):
    age_group_cost = df.groupby('Age_Group')['Billing Amount'].agg(["sum","max","min","count"]).rename(
    columns = {"sum":"Total_Billing","max":"Highest Billing","min":"Cheapest Billing","count":"Age group count"})
    age_group_cost
    
    #Sort and reset index
    age_group_statistic = age_group_cost.reset_index().sort_values(by = ['Total_Billing','Highest Billing','Cheapest Billing',"Age group count"],ascending = False)
                                                                      
    return  age_group_statistic 
    
age_group_statistic = age_group_stat(df)
age_group_statistic                                                                                                                                     
```

##### Data visualization for age group billing cost.

```python
fig, axes = plt.subplots(nrows=4, figsize=(20, 25))

#Create barplot to show Total billing by age group.
sns.barplot(data=age_group_statistic, y="Age_Group", x="Total_Billing", ax=axes[0])
axes[0].set_title('Total Billing by Age Group') 

#Create barplot to show Highest billing amount by age group.
sns.barplot(data=age_group_statistic, y="Age_Group", x="Highest Billing", ax=axes[1])
axes[1].set_title('Highest Billing by Age Group')

#Create barplot to show cheapest billing amount by age group.
sns.barplot(data=age_group_statistic, y="Age_Group", x="Cheapest Billing", ax=axes[2])
axes[2].set_title('Cheapest Billing by Age Group') 

#Create barplot to show age group count.
sns.barplot(data=age_group_statistic, y="Age_Group", x="Age group count", ax=axes[3])
axes[3].set_title('Age Group Count')

plt.tight_layout()
plt.show()
    
    
```

#### 2.What was the Medical Condition with the highest billing cost?

```python
def medical_medication_stat(df):
    medical_condition = df.groupby(['Medical Condition'])["Billing Amount"].agg(['sum','count']).reset_index().sort_values(by= ['sum','count'],ascending = False).head(10)
    medical_condition
    medical_condition_cost = medical_condition.rename(columns ={"sum":"Total_billing","count":"Patient_count"})
    return medical_condition_cost

medical_medication_statistic = medical_medication_stat(df)
medical_medication_statistic
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=2, figsize=(20, 15))
# Plot Total Billing by Medical Condition
sns.barplot(data=medical_medication_statistic, x="Medical Condition", y="Total_billing", ax=axes[0])
axes[0].set_title('Total Billing by Medical Condition')
# Add labels to the bars
for bar in axes[0].patches:
    axes[0].text( bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():,.2f}', ha='center', va='bottom' )
# Plot Number of Medical Condition
sns.barplot(data=medical_medication_statistic, x="Medical Condition", y="Patient_count", ax=axes[1])
axes[1].set_title('Number of Medical Condition')
# Add labels to the bars
for bar in axes[1].patches:
    axes[1].text( bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():,.0f}', ha='center', va='bottom' )
```

#### 4.Find admission type with the most patient intake, Displaying their respective billing cost.

```python
def admission_type_stas(df):
    admission_type = df.groupby('Admission Type')['Billing Amount'].agg(['sum','count']).reset_index().sort_values(
    by = ['count','sum'],ascending = False)
    #Rename agg columns
    admission_type_intake = admission_type.rename(columns = {'count':'admission_type_count','sum':'Total_billing_cost'})
    return admission_type_intake
admission_type_stastistic = admission_type_stas(df)
admission_type_stastistic
```

#### 6. What was the percentage ratio of gender?

```python
def calculate_gender_statistic(df):

    gender_counts = df['Gender'].value_counts()
    gender_percentage = (gender_counts / gender_counts.sum()) * 100

    # Combine into a single DataFrame
    gender_statistics = pd.DataFrame({
    'Gender_Count': gender_counts,
    'Percentage (%)': gender_percentage
     })

    return gender_statistics

statistic = calculate_gender_statistic(df)
print(statistic)
```

##### Data Visualization for percentage ratio of gender.

```python
#Create a piechart.
fig, axes = plt.subplots(figsize=(12, 7)) 

#Pie chart for percentages
axes.pie(statistic['Percentage (%)'], labels=statistic.index,autopct='%.2f%%', startangle=180)

#Add title
axes.set_title("Gender_distribution")
plt.tight_layout() 
#Show plot
plt.show()
```

#### 7. Find the correlation relationship between Age, Billing Amount , and Room Number.

#### 9. Find the top 10 patients with the highest billing cost.

```python
def patient_billing_cost(df):
    patient_billing_cost = df.groupby('Name')['Billing Amount'].sum()

    #sort the value in largest order of 10.
    billing_cost_patient = patient_billing_cost.sort_values(ascending=False).nlargest(10).reset_index()
    return  billing_cost_patient
top_ten_patient = patient_billing_cost(df)
print(top_ten_patient)

```

##### Data visualizaton with patient with billing amount


```python
plt.figure(figsize=(15,8))
sns.barplot(top_ten_patient,y = "Name", x = "Billing Amount")
plt.title("Top ten patient billing cost")
plt.tight_layout()
plt.show()
```

#### 10. Find the top 10 patients with the least billing amount.

```python
def top_ten_smallest_billing_amount(df):
    billing_cost_patient = df.groupby('Name')['Billing Amount'].sum()

    #sort the value in smallest order of 10.
    billing_cost_patient = billing_cost_patient.sort_values(ascending=True).nsmallest(10).reset_index()
    billing_cost_patient
    return  billing_cost_patient
top_ten_least_patient_billing_cost = top_ten_smallest_billing_amount(df)
print(top_ten_least_patient_billing_cost)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(top_ten_least_patient_billing_cost,y = "Name", x = "Billing Amount")
plt.title("Top ten patient with the least billing cost")
plt.tight_layout()
plt.show()
plt.tight_layout() 
```

  #### 11.What year had the most patient admission intake?

```python
#Create Admission year intake column.
df['Admission Year'] = df["Date of Admission"].dt.year
df['Admission Year']
```

```python
def year_stats(df):
    patient_year_stat = df.groupby('Admission Year')['Billing Amount'].agg(['count','sum']).rename(
    columns = {'count':'Number of patient admission','sum':'Total billing amount'}).reset_index()
    patient_year_stat
    return patient_year_stat
patient_year_statistic = year_stats(df)
patient_year_statistic
```

##### Data visualization for patient yearly admission trend.

```python
fig, axes = plt.subplots(nrows=2, figsize=(15, 10))

#Create lineplot.
sns.lineplot(patient_year_statistic, y = "Number of patient admission", x = "Admission Year",marker = 'o', ax = axes[0])
axes[0].set_title('Yearly trend by patient admission') 
axes[0].grid(True)



#Ceate barplot.
sns.barplot(patient_year_statistic, y = "Total billing amount", x = "Admission Year",ax = axes[1])
axes[1].set_title('Total billing amount by year of admitted admission')

plt.tight_layout()
plt.show()
```

##### 12. What year had the most patient discharge?

```python
#Create Admission year intake column.
df['Discharge Year'] = df["Discharge Date"].dt.year
df['Discharge Year']
```

```python
def yearly_discharged(df):
    patient_discharged_stat = df.groupby('Discharge Year')['Billing Amount'].agg(['count','sum']).rename(
    columns = {'count':'Number of patient discharged','sum':'Total billing amount'}).reset_index()
    patient_discharged_stat
    return patient_discharged_stat
patient_yearly_discharged = yearly_discharged(df)
patient_yearly_discharged 
```

##### Data visualization for patient yearly discharged trend.

```python
fig, axes = plt.subplots(nrows=2, figsize=(15, 10))

#create a lineplot
sns.lineplot(patient_yearly_discharged , y = "Number of patient discharged", x = "Discharge Year", marker = 'o', ax = axes[0])
axes[0].set_title('Yearly trend by number discharge patient') 
axes[0].grid(True)


#create a barplot
sns.barplot(patient_yearly_discharged , y = "Total billing amount", x = "Discharge Year", ax = axes[1])
axes[1].set_title('Total billing amount by year of discharge patient')


plt.tight_layout()
plt.show()
```

#### 13. Find top five doctor performances analysis.

```python
def doctor_analysis_performances(df):
    analysis = df.groupby('Doctor')['Billing Amount'].agg(['count','sum']).rename(
    columns = {'count':'count of treated patient','sum':'Total billing amount'}).reset_index().sort_values(['count of treated patient',
    'Total billing amount'],ascending = False)
    analysis.head(5)
    return analysis.head(5)
doctor_analysis = doctor_analysis_performances(df)
doctor_analysis
```

```python
#Create barplot
plt.figure(figsize=(15,5))
sns.barplot(doctor_analysis, x = 'Doctor', y = 'Total billing amount')
plt.title('Top five billing amount by doctor')
plt.xticks(rotation = 45)
plt.show()
```
